## Question 1: Bayesian Inferencing

(a) Assume that training data points $\mathbf{X}=[\mathbf{x}_1,\ldots,\mathbf{x}_N]\in\mathbb{R}^{d\times N}$ and corresponding scalar target values $\mathbf{t}=[t_1,\ldots,t_N]^\text{T}$ are drawn independently with a likelihood in the form 
$$\begin{aligned} p(\mathbf{t}\vert\mathbf{X},\mathbf{w},\beta)=\prod_{n=1}^N\mathcal{N}(t_n\vert\mathbf{w}^\text{T}\mathbf{\phi}(\mathbf{x}_n), \beta^{-1})
\end{aligned}$$ 
for parameters $\mathbf{w}$ and $\beta$. Let a Gaussian prior over the coefficients $\mathbf{w}$ be given by
$$\begin{aligned}
        p(\mathbf{w}|\alpha)=\mathcal{N}(\mathbf{w}\vert\mathbf{0},\alpha^{-1}\mathbf{I}).
\end{aligned}$$ 
Under these assumptions it can be shown that the resulting posterior distribution 

$$\begin{aligned}
            p(\mathbf{w}|\mathbf{X},\mathbf{t},\beta,\alpha)\propto p(\mathbf{t}\vert\mathbf{X},\mathbf{w},\beta) p(\mathbf{w}|\alpha)
\end{aligned}$$ 
is again a Gaussian, wherefore the prior and posterior distributions are called *conjugate*. Show that the mean $\pmb{\mu}_N$ and variance $\mathbf{\Sigma}_N$ of the posterior distribution are given by 

$$\begin{aligned}
    \pmb{\mu}_N=\beta\mathbf{\Sigma}_N\mathbf{\Phi}^\text{T}\mathbf{t}\quad\text{and}\quad\mathbf{\Sigma}_N^{-1}=\alpha\mathbf{I}+\beta\mathbf{\Phi}^\text{T}\mathbf{\Phi}.
\end{aligned}$$

**Solution:**

$$\begin{aligned} 
    \mathcal{N}(\mathbf{w}\vert\pmb{\mu}_N,\mathbf{\Sigma}_N)=((2\pi)^d\vert\mathbf{\Sigma}_N\vert)^{-\frac{1}{2}}
    \exp{(-\frac{1}{2}(\mathbf{w}-\pmb{\mu}_N)^\text{T}\mathbf{\Sigma}_N^{-1}(\mathbf{w}-\pmb{\mu}_N))}
\end{aligned}$$
<br>
&emsp;&emsp;The term in the exponential:
$$\begin{aligned}
    -\frac{1}{2}(\mathbf{w}-\pmb{\mu}_N)^\text{T}\mathbf{\Sigma}_N^{-1}(\mathbf{w}-\pmb{\mu}_N)
    =-\frac{1}{2}\mathbf{w}^\text{T}\mathbf{\Sigma}_N^{-1}\mathbf{w} 
     -\pmb{\mu}_N^\text{T}\mathbf{\Sigma}_N^{-1}\mathbf{w} + constant \qquad\qquad(1)
\end{aligned}$$
<br>
&emsp;&emsp;The term in the exponential of posterior:
$$\begin{aligned} 
    -\sum_{n=1}^N\frac{1}{2}(\mathbf{t}_n-\mathbf{w}^\text{T}\mathbf{\phi}(\mathbf{x}_n))^\text{T}
    \beta(\mathbf{t}_n-\mathbf{w}^\text{T}\mathbf{\phi}(\mathbf{x}_n))
    -\frac{1}{2}(\mathbf{w}-0)^\text{T}\alpha\mathbf{I}(\mathbf{w}-0)
\end{aligned}$$
<br>
$$\begin{aligned}
    =-\frac{\beta}{2}\sum_{n=1}^N(\mathbf{t}_n-\mathbf{w}^\text{T}\mathbf{\phi}(\mathbf{x}_n))^2
    -\frac{\alpha}{2}\mathbf{w}^\text{T}\mathbf{w}
\end{aligned}$$
<br>
$$\begin{aligned}
    =-\frac{\beta}{2}\sum_{n=1}^N(\mathbf{t}_n^2-2\mathbf{t}_n\mathbf{w}^\text{T}\mathbf{\phi}(\mathbf{x}_n)+\mathbf{w}^\text{T}\mathbf{\phi}(\mathbf{x}_n)\mathbf{\phi}(\mathbf{x}_n)^\text{T}\mathbf{w})
    -\frac{\alpha}{2}\mathbf{w}^\text{T}\mathbf{w}
\end{aligned}$$
<br>
$$\begin{aligned}

    \beta\mathbf{\phi}(\mathbf{x}_n)\mathbf{\phi}(\mathbf{x}_n)^\text{T}+\alpha\mathbf{I})\mathbf{w}

\end{aligned}$$

$$\begin{aligned}

\end{aligned}$$

## Question 2: Loss Functions For Regression

Consider the expected loss

$$\begin{aligned}
        \mathbb{E}[L_q]=\int_{\mathbb{R}^d}\int_\mathbb{R}L_q(t,y(\mathbf{x}))p(\mathbf{x},t)\text{d}t\text{d}\mathbf{x}
\end{aligned}$$ 
for the *Minkowski loss* function $L_q(t,y(\mathbf{x}))=\vert y(\mathbf{x})-t\vert^q$. In the lecture you have seen that the optimal regression function $y:\mathbb{R}^d\rightarrow\mathbb{R}$ that minimizes $\mathbb{E}[L_q]$ for the squared loss $q=2$ is given by the *conditional mean* $y(\mathbf{x})=\mathbb{E}[t\vert\mathbf{x}]=\int_\mathbb{R}tp(t\vert\mathbf{x})\text{d}t$.

(a) Show that for $q=1$ the optimal regression function is given by the *conditional median* 
$$\begin{aligned}
    y(\mathbf{x})=\mathop{\mathrm{argmin}}_{\hat y}\int_\mathbb{R}\vert\hat y -t\vert p(t\vert\mathbf{x})\text{d}t
\end{aligned}$$

**Solution:**


$$\begin{aligned}
    \mathbb{E}[L_q]=\int_{\mathbb{R}^d}\int_\mathbb{R}\vert\hat y -t\vert^\text{q}
    p(\mathbf{x},t)\text{d}t\text{d}\mathbf{x}
\end{aligned}$$
<br>
$$\begin{aligned}
    =\int_{\mathbb{R}^d} p(\mathbf{x})[\int_\mathbb{R}\vert\hat y -t\vert^\text{q}
    p(t\vert\mathbf{x})\text{d}t]\text{d}\mathbf{x}
\end{aligned}$$
<br>
&emsp;&emsp;For every $\mathbf{x}\in\mathbb{R}^d$, choose $\hat y = y(\mathbf{x})\in\mathbb{R}$, such that
<br>
$$\begin{aligned}
    \mathbf{F}_{q}(\hat y\vert\mathbf{x})=\int_{\mathbb{R}}\vert\hat y -t\vert^\text{q}
    p(t\vert\mathbf{x})\text{d}t
\end{aligned}$$
<br>
&emsp;&emsp;For q=1, this gives 
<br>
$$\begin{aligned}
    y(\mathbf{x})=\mathop{\mathrm{argmin}}_{\hat y \in\mathbb{R} }[\mathbf{F}_{q}(\hat y\vert\mathbf{x})]
\end{aligned}$$


(b) Show that for $q\rightarrow 0$ the optimal regression function is given
by the *conditional mode* 

$$\begin{aligned}
    y(\mathbf{x})=\mathop{\mathrm{argmax}}_{t\in\mathbb{R}}p(t\vert\mathbf{x})
\end{aligned}$$

**Solution:**


&emsp;&emsp;Assume: $p(•\vert\mathbf{x})$ is  uniformly continuous
<br>
<br>
&emsp;&emsp;Fix $0<\varepsilon<<1$:
$$\begin{aligned}
    \exists0<\delta=\delta(\varepsilon)<<1:\quad\forall\hat y\in\mathbb{R},t\in\mathbf{U}_{\delta}(0)
\end{aligned}$$
<br>
$$\begin{aligned}
    \vert p(\hat y + t\vert\mathbf{x})- p(\hat y \vert\mathbf{x})\vert < \varepsilon
\end{aligned}$$
<br>
&emsp;&emsp;Consider:
$$\begin{aligned}
    \mathbf{F}_{q}(\hat y\vert\mathbf{x})=\int_{\mathbb{R}}\vert\hat y -t\vert^\text{q}
    p(t\vert\mathbf{x})\text{d}t
    =\int_{\mathbb{R}}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t
\end{aligned}$$
<br>
$$\begin{aligned}
=\int_{\mathbb{R}\setminus\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t
    +\int_{\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t
    \qquad (1)
\end{aligned}$$
<br>
<br>
&emsp;&emsp;The former in Eqn.(1): $\quad$ Choose $0< q=q(\delta) <<1$
<br>
$$\begin{aligned}
    \int_{\mathbb{R}\setminus\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t
    -\int_{\mathbb{R}\setminus\mathbf{U}_{\delta}(0)}p(\hat y + t\vert\mathbf{x})\text{d}t\quad<\varepsilon
\end{aligned}$$
<br>
$$\begin{aligned}
    \int_{\mathbb{R}\setminus\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t
    =\int_{\mathbb{R}\setminus\mathbf{U}_{\delta}(0)}p(\hat y + t\vert\mathbf{x})\text{d}t+\mathbb{O}(\varepsilon)
\end{aligned}$$
<br>
&emsp;&emsp;Eqn.(1):
$$\begin{aligned}
    \int_{\mathbb{R}\setminus\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t
    +\int_{\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t
\end{aligned}$$
<br>
$$\begin{aligned}
=\int_{\mathbb{R}\setminus\mathbf{U}_{\delta}(0)}p(\hat y + t\vert\mathbf{x})\text{d}t
    +\int_{\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t+\mathbb{O}(\varepsilon)
\end{aligned}$$
<br>
$$\begin{aligned}
    =\int_{\mathbb{R}}p(\hat y + t\vert\mathbf{x})\text{d}t
    +\int_{\mathbf{U}_{\delta}(0)}\vert t \vert^{q}p(\hat y + t\vert\mathbf{x})\text{d}t+\mathbb{O}(\varepsilon)
\end{aligned}$$
<br>
$$\begin{aligned}
    =1-\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})p(\hat y + t\vert\mathbf{x})\text{d}t+\mathbb{O}(\varepsilon)\qquad(2)
\end{aligned}$$
<br>
&emsp;&emsp;The middle term in Eqn.(2):
$$\begin{aligned}
    \int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})p(\hat y + t\vert\mathbf{x})\text{d}t
    =\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})p(\hat y + t\vert\mathbf{x})\text{d}t
    -\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})p(\hat y \vert\mathbf{x})\text{d}t
\end{aligned}$$
<br>
$$\begin{aligned}
    =\vert\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})
    [p(\hat y + t\vert\mathbf{x})-p(\hat y \vert\mathbf{x})]\text{d}t\vert
\end{aligned}$$
<br>
$$\begin{aligned}
    <\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})\vert 
    p(\hat y + t\vert\mathbf{x})-p(\hat y +\mathbf{x})\vert\text{d}t
\end{aligned}$$
<br>
&emsp;&emsp;Since $\vert p(\hat y + t\vert\mathbf{x})-p(\hat y +\mathbf{x})\vert<\varepsilon$ 
            by uniform continuity,
$$\begin{aligned}
    <\varepsilon\int_{\mathbf{U}_{\delta}(0)}1-\vert t\vert^{q}\text{d}t=\mathbb{O}(\varepsilon)
\end{aligned}$$
<br>
&emsp;&emsp;Then,
$$\begin{aligned}
    \int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q}) p(\hat y + t\vert\mathbf{x})\text{d}t
    =\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q}) p(\hat y \vert\mathbf{x})\text{d}t+\mathbb{O}(\varepsilon)
    =[\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})\text{d}t] p(\hat y \vert\mathbf{x})+\mathbb{O}(\varepsilon)
\end{aligned}$$
<br>
&emsp;&emsp;Back to Eqn.(2):
$$\begin{aligned}
    1-\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})p(\hat y + t\vert\mathbf{x})\text{d}t
    +\mathbb{O}(\varepsilon)
    =1-[\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})\text{d}t] p(\hat y \vert\mathbf{x})
    +\mathbb{O}(\varepsilon)
\end{aligned}$$
<br>
&emsp;&emsp;Due to $\int_{\mathbf{U}_{\delta}(0)}(1-\vert t \vert^{q})\text{d}t>0$,
$$\begin{aligned}
    \mathop{\mathrm{argmin}}_{\hat y \in\mathbb{R}}[\mathbf{F}_{q}(\hat y\vert\mathbf{x})]
    =\mathop{\mathrm{argmax}}_{\hat y \in\mathbb{R}}[p(\hat y \vert\mathbf{x})]=y(\mathbf{x})
\end{aligned}$$

## Question 3: Least Squares Regression

In this problem your task is to learn a function, $f: \mathbb{R} \rightarrow \mathbb{R}$, given the training data using least squares regression.

You will use the datasets `test.txt` and `train.txt`, in which the data matrices contain 1D data points $x_i \in \mathbb{R}$ in the first column and the corresponding target output values $y_i \in \mathbb{R}$ in the second column.

Instead of the polynomial basis functions discussed in class, here you will use the Fourier basis functions: $$\begin{aligned}
        \phi_0(x) = 1
        \hspace{1cm}  \phi_{2l-1}(x) = \frac{1}{l} \text{cos}(2\pi l x)
        \hspace{1cm}  \phi_{2l}(x) = \frac{1}{l} \text{sin}(2\pi l x)
\end{aligned}$$ 
where $l\ (l = 1,2,3,\ldots)$ is the frequency of the basis function. 

Thus, given the maximal frequency $k$, we will have $2k+1$ basis functions. 

For a fixed maximal frequency $k$, every data pair $(x_i, y_i),\ i=1,\ldots,N,$ defines an equation $$\begin{aligned}
        w_0\cdot\phi_0(x_i)+w_1\cdot\phi_1(x_i)+\cdots+w_{2k}\cdot\phi_{2k}(x_i)=y_i
    \end{aligned}$$ with coefficients $w_0,\ldots,w_{2k}\in\mathbb{R}$.
    
Writing them into a single matrix equation, we get 
$$\begin{aligned}
        \underbrace{\begin{bmatrix}
            \phi_0(x_0) & \phi_1(x_0) & \cdots & \phi_{2k}(x_0) \\
            \phi_0(x_1) & \phi_1(x_1) & \cdots & \phi_{2k}(x_1) \\
            \vdots & \vdots & \vdots & \vdots\\
            \phi_0(x_N) & \phi_1(x_N) & \cdots & \phi_{2k}(x_N)
        \end{bmatrix}}_{\mathbf{\Phi}}
        \cdot
        \underbrace{\begin{bmatrix}
            w_0\\
            w_1\\
            \vdots\\
            w_{2k}
        \end{bmatrix}}_{\mathbf{w}}
        =
        \underbrace{\begin{bmatrix}
            y_0\\
            y_1\\
            \vdots\\
            y_N
        \end{bmatrix}}_{\mathbf{y}}.
\end{aligned}$$ 
Since the system $\mathbf{\Phi}\cdot\mathbf{w}=\mathbf{y}$ is usually overconstrained, it does not have an exact solution.

Instead, we consider the solution to the minimization problem 
$$\begin{aligned}
        \lVert\mathbf{\Phi}\cdot\mathbf{w}-\mathbf{y}\rVert^2\rightarrow\min,
\end{aligned} $$
which is given by the *normal equations* $\mathbf{w}=(\mathbf{\Phi}^\text{T}\mathbf{\Phi})^{-1}\mathbf{\Phi}^\text{T}\mathbf{y}$.

````python
def fourier_features(x, n):
    """
    Produces fourier features for the input.

    Parameters
    ----------
    x : array_like
        input array of shape (N,).
    n : int
        feature dimensionality.

    Returns
    -------
    x_tilda : array_like
        resulting fourier features of shape (N,).

    """
    # read the task
    # YOUR CODE HERE
    
    if n == 0:
        return np.ones_like(x)
    if n % 2 == 1:
        l = (n+1) / 2
        return (1/l) * np.cos(2 * np.pi * l * x)
    else:
        l = n / 2
        return (1/l) * np.sin(2 * np.pi * l * x)
    
    raise NotImplementedError()
````

### Question 3.b: Feature matrix, calculation of weights and metrics. Regression.
For each $k$, compute the *root mean square error* $E_\text{RMS}=\sqrt{E(\mathbf{w})^2/N}$ on both, training and test data, as a function of maximum frequency $k$.

````python
def get_feature_matrix(train, max_k):
    """
    Get a feature matrix.

    Parameters
    ----------
    train : array_like
        input array of shape (N,).
    max_k : int
        maximum number of fourier features K.

    Returns
    -------
    features : array_like
        resulting features matrix (N,K).

    """
    # YOUR CODE HERE
    features = []
    for i in range(max_k):
        # np.newaxis: 1D->2D, ex. x=np.array([1,2,3,4]), x[:, np.newaxis]: from (4,) to (4,1)
        features.append(fourier_features(train, i)[:, np.newaxis])
    features = np.hstack(features)
    return features

    raise NotImplementedError()
````

````python
assert 1 == get_feature_matrix(np.array([0]), 1)
assert np.allclose(np.array([[1, 1, 0]]), get_feature_matrix(np.array([0]), 3))
````

````python
def calculate_weights(train_features, train_labels):
    """
    Weights calculation for regression.

    Parameters
    ----------
    features : array_like
        input array of shape (N,K).
    labels : array_like
        labels of shape (N).

    Returns
    -------
    w : array_like
        resulting weights matrix (N,).

    """
    # hint: you can use np.linalg.pinv for this one
    # YOUR CODE HERE
    w = np.linalg.pinv(train_features) @ train_labels
    return w

    raise NotImplementedError()
````

````python
assert 1 == calculate_weights(np.array([[1]]), np.array([1]))
assert np.allclose(
    [0.5, 0.5], calculate_weights(np.array([[0, 0], [1, 1]]), np.array([0, 1]))
)
````

````python
def get_rmse(labels, prediction):
    """
    Calculate root mean square metrics.

    Parameters
    ----------
    labels : array_like
        labels of shape (N,).
    prediction : array_like
        prediction array of shape (N,).

    Returns
    -------
    rmse : float
        resulting rmse.

    """
    # YOUR CODE HERE
    rmse = np.sqrt(np.mean(np.square(labels - prediction)))
    return rmse

    raise NotImplementedError()
````

````python
assert 0 == get_rmse(np.array([1, 1]), np.array([1, 1]))
assert 1 == get_rmse(np.array([0, 0]), np.array([1, 1]))
````

````python
# you don't have to modify this cell
x = np.linspace(0, 1, 1000)
train_errors = []
test_errors = []

for max_k in range(1, 18, 2):
    # calculate features and weights
    phi = get_feature_matrix(train_X, max_k)
    w = calculate_weights(phi, train_y)

    # get predictions
    train_prediction = get_feature_matrix(train_X, max_k) @ w
    test_prediction = get_feature_matrix(test_X, max_k) @ w
    prediction_on_x = get_feature_matrix(x, max_k) @ w

    # calculate rmse metrics
    train_rmse = get_rmse(train_y, train_prediction)
    test_rmse = get_rmse(test_y, test_prediction)
    train_errors.append(train_rmse)
    test_errors.append(test_rmse)

    # plotting
    plt.title(f"Train RMSE = {train_rmse:.2f}, Test RMSE = {test_rmse:.2f}")

    legend = []
    legend.append("test")
    plt.plot(test_X, test_y, "bo", markersize=0.5)
    legend.append("train")
    plt.plot(train_X, train_y, "rx")
    legend.append(f"Model with k = {max_k}")
    plt.plot(x, prediction_on_x, "g-")

    plt.xlim([0, 1])
    plt.ylim([-2, 2])
    plt.legend(legend)
    plt.show()
````

### Question 3.c: Ridge regression
Extend your code to incorporate a prior on the weights of the regression function to perform *Ridge Regression*.
Repeat the above tasks with various regularization weights $\lambda$.

````python
def calculate_weights_ridge_regression(train_features, train_labels, alpha=0.01):
    """
    Simple weights calculation from features and labels, with ridge regression.

    Parameters
    ----------
    features : array_like
        input array of shape (N,K).
    labels : array_like
        labels of shape (N).

    Returns
    -------
    w : array_like
        resulting weights matrix (N,).

    """
    # YOUR CODE HERE
    phi_inv = train_features.T @ train_features
    w = np.linalg.pinv(phi_inv + alpha * np.eye(phi_inv.shape[0], phi_inv.shape[1])) @ train_features.T @ train_labels
         
    return w

    raise NotImplementedError()
````

### Question 4.a: Ridge regression with exponential kernel
Squared exponential kernel: 
$$\begin{aligned}
                k(x,y) = \exp\left( - \frac{( x - y )^2}{2l^2} \right)
\end{aligned}$$

````python
def exponential_kernel(x, y, length_scale=0.4):
    """
    Calculation of exponential kernel based on inputs x and y.

    Parameters
    ----------
    x : array_like
        input array of shape (N).
    y : array_like
        labels of shape (N).
    length_scale : float
        parameter of exponential kernel.

    Returns
    -------
    kernel : array_like
        resulting kernel matrix (N,N).

    """
    # YOUR CODE HERE
    x, y = x[:, None], y[:, None]
    
    exp_inner = (((-2 * x @ y.T) / (length_scale ** 2))
                + np.sum(y ** 2 / (length_scale ** 2), axis=1)
                + np.sum(x ** 2 / (length_scale ** 2), axis=1)[:, None])
    kernel = np.exp(-0.5 * exp_inner)
    return kernel

    raise NotImplementedError()
````

````python
def get_matrix_a(kernel, train_labels, alpha=0.1):
    """
    Calculation of representation matrix.

    Parameters
    ----------
    kernel : array_like
        input array of shape (N,K).
    labels : array_like
        labels of shape (N).

    Returns
    -------
    a : array_like
        resulting solution to dual representation matrix (N,).

    """
    #YOUR CODE HERE
    a = np.linalg.inv(kernel + alpha * np.eye(kernel.shape[0])) @ train_labels
    
    return a

    raise NotImplementedError()
````

````python
def estimate_at_location(test, train, a, kernel_function, kernel_parameters):
    """
    Simple weights calculation from features and labels, with ridge regression.

    Parameters
    ----------
    test : array_like
        input test data of shape (M,);
    train : array_like
        input train data of shape (N,);
    a : array_like
        dual representation matrix (N,);
    kernel_function : Callable
        desired kernel funciton. Default: polynomial_kernel;
    kernel_parameters : float
        kernel paramters for `kernel_function`.

    Returns
    -------
    prediction : array_like
        resulting estimation (M,).

    """
    # YOUR CODE HERE
    kernel = kernel_function(test, train, kernel_parameters)
    prediction = kernel @ a
    return prediction

    raise NotImplementedError()
````

````python
# you don't have to modify this cell
x = np.linspace(0, 1, 1000)
train_errors = []
test_errors = []

for kernel_parameters in np.linspace(0.3, 1.5, 5):
    # compute kernel
    kernel_matrix = exponential_kernel(train_X, train_X, length_scale=kernel_parameters)
    # compute solution for dual representation
    a = get_matrix_a(kernel_matrix, train_y, 0.001)

    # compute estimate at sampled locations
    train_prediction = estimate_at_location(
        train_X,
        train_X,
        a,
        kernel_function=exponential_kernel,
        kernel_parameters=kernel_parameters,
    )
    test_prediction = estimate_at_location(
        test_X,
        train_X,
        a,
        kernel_function=exponential_kernel,
        kernel_parameters=kernel_parameters,
    )
    prediction = estimate_at_location(
        x,
        train_X,
        a,
        kernel_function=exponential_kernel,
        kernel_parameters=kernel_parameters,
    )

    # calculate rmse metrics
    train_rmse = get_rmse(train_y, train_prediction.ravel())
    test_rmse = get_rmse(test_y, test_prediction.ravel())
    train_errors.append(train_rmse)
    test_errors.append(test_rmse)

    # plotting
    plt.title(f"Train RMSE = {train_rmse:.2f}, Test RMSE = {test_rmse:.2f}")

    legend = []
    legend.append("test")
    plt.plot(test_X, test_y, "bo", markersize=0.5)
    legend.append("train")
    plt.plot(train_X, train_y, "rx")
    legend.append(f"Model with k = {kernel_parameters}")
    plt.plot(x, prediction, "g-")

    plt.xlim([0, 1])
    plt.ylim([-2, 2])
    plt.legend(legend)
    plt.show()
````
